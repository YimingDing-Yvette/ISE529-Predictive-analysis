```python
data = pd.read_csv("stocks.csv",parse_dates = ["Date"]) #parse_dates: telling python this is date,type is datetime
data[-4:]
```

```python
Date_col = data.pop("Date") # drop the date feature
data.index = Date_col # extract the date as index, not the number previously
data[:5]
```

```python
data.AAPL.max()
```

```python
data.loc[data["AAPL"] == 228.523819]
```

#### answer 2a): At October 3rd,2018, AAPL stock was at highest price

### 2b)

```python
net_returns = data/data.shift(1) - 1
net_returns[:5]
daily_pc = net_returns.copy()
daily_pc.index = Date_col
daily_pc[:5]
```

```python
gross_returns = (1+ net_returns).cumprod() # cumulative productivity
gross_returns[:5]
daily_cr = gross_returns.copy()
daily_cr.plot(figsize = (12,6)) # length by width
plt.title("All cumulative returns")
plt.grid()
plt.legend(loc = 2) # location of legend(label)
```

```python
daily_cr[-5:]
```

```python
wealth = daily_cr.AAPL[-1] * 100
wealth
```

#### Answer 2b): The wealth would be 195.69 if i had invested 100 dollars in Jan 2015 in AAPL.
